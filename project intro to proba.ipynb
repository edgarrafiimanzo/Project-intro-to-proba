## Uji Hipotesis

### 1. Tagihan kesehatan perokok lebih tinggi daripada tagihan kesehatan non perokok

Maka hipotesis yang akan kita uji adalah:
$$H_0: \mu_{Tagihan perokok} \leq \mu_{Tagihan non perokok}$$
$$H_1: \mu_{Tagihan perokok} > \mu_{Tagihan non perokok}$$

```python
tagihan_perokok=df['charges'][df['smoker']=='yes']
mean_tagihan_perokok = tagihan_perokok.mean()
mean_tagihan_perokok
```

```python
tagihan_non=df['charges'][df['smoker']=='no']
mean_tagihan_non = tagihan_non.mean()
mean_tagihan_non
```

```python
alpha = 0.05
```

```python
np.var(tagihan_perokok), np.var(tagihan_non)
```

```python
from scipy.stats import ttest_ind
```

```python
stat, p = ttest_ind(a = tagihan_perokok , b = tagihan_non , equal_var = False , alternative = 'greater' )
print(stat, p)
```

```python
if p > alpha:
    print('Tidak cukup bukti untuk menolak klaim H_0')
else:
    print('Tagihan kesehatan perokok lebih tinggi dari non perokok')
```

### 2. Tagihan kesehatan nasabah dengan BMI diatas 25 lebih tinggi daripada tagihan kesehatan nasabah dengan BMI dibawah 25

Maka hipotesis yang akan kita uji adalah:
$$H_0: \mu_{Tagihan BMI >25} \leq \mu_{Tagihan BMI <25}$$
$$H_1: \mu_{Tagihan BMI >25} > \mu_{Tagihan BMI <25}$$

```python
tagihan_25plus = df['charges'][df['bmi']>25]
tagihan_25plus
```

```python
tagihan_25minus = df['charges'][df['bmi']<25]
tagihan_25minus
```

```python
alpha_1 = 0.05
```

```python
np.var(tagihan_25plus), np.var(tagihan_25minus)
```

```python
stat_1, p_1 = ttest_ind(a = tagihan_25plus , b = tagihan_25minus , equal_var = False , alternative = 'greater' )

print(stat_1, p_1)
```

```python
if p_1 > alpha_1:
    print('Tidak cukup bukti untuk menolak klaim H_0')
else:
    print('Tagihan kesehatan nasabah dengan BMI >25 lebih tinggi dari nasabah dengan BMI <25')
```

### 3. BMI laki-laki dan perempuan sama

Maka hipotesis yang akan kita uji adalah:
$$H_0: \mu_{BMI laki-laki} = \mu_{BMI perempuan}$$
$$H_1: \mu_{BMI laki-laki} \not= \mu_{BMI perempuan}$$

```python
bmi_laki = df['charges'][df['sex']=='male']
bmi_wanita = df['charges'][df['sex']=='female']
```

```python
print(np.var(bmi_laki), np.var(bmi_wanita))
alpha_2 = 0.05
```

```python
stat_2, p_2 = ttest_ind(a = bmi_laki , b = bmi_wanita , equal_var = False , alternative = 'two-sided' )
print(stat_2, p_2)
```

```python
if p_2 > alpha_2:
    print('Pada tingkat signifikansi 0.05 tidak cukup bukti untuk menolak klaim bahwa BMI laki-laki sama dengan BMI perempuan')
else:
    print('BMI laki-laki tidak sama dengan BMI perempuan')
```
